```python
categorical_cols = [
    "device_type",
    "user_type",
    "marketing_channel",
    "product_category",
    "payment_method",
    "visit_weekday",
    "visit_season",
    "session_duration_bucket"
]

for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```

```python
numeric_cols = [
    "unit_price",
    "quantity",
    "discount_percent",
    "discount_amount",
    "revenue",
    "pages_viewed",
    "time_on_site_sec",
    "rating"
]

(df[numeric_cols] < 0).sum()
```

```python
df["visit_date"] = pd.to_datetime(df["visit_date"])

df.info()
```

```python
print(df.shape)

df.head()
```

```python
df.to_csv("../data/cleaned/indian_ecommerce_cleaned.csv", index=False)
```
