```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("../data/cleaned/indian_ecommerce_cleaned.csv")

df["visit_date"] = pd.to_datetime(df["visit_date"])

if "visit_day" not in df.columns:
    df["visit_day"] = df["visit_date"].dt.day

if "visit_month" not in df.columns:
    df["visit_month"] = df["visit_date"].dt.month

if "visit_weekday" not in df.columns:
    df["visit_weekday"] = df["visit_date"].dt.day_name()

if "visit_season" not in df.columns:

    def season(month):
        if month in [12,1,2]:
            return "Winter"
        elif month in [3,4,5]:
            return "Summer"
        elif month in [6,7,8,9]:
            return "Monsoon"
        else:
            return "Autumn"

    df["visit_season"] = df["visit_month"].apply(season)

if "session_duration_bucket" not in df.columns:

    df["session_duration_bucket"] = pd.cut(
        df["time_on_site_sec"],
        bins=[0,120,300,600,100000],
        labels=["Very Short","Short","Medium","Long"]
    )

if "revenue_normalized" not in df.columns:

    scaler = MinMaxScaler()

    df["revenue_normalized"] = scaler.fit_transform(
        df[["revenue"]]
    )

print("Feature Engineered Columns")

columns = [
    "visit_day",
    "visit_month",
    "visit_weekday",
    "visit_season",
    "session_duration_bucket",
    "revenue_normalized"
]

for col in columns:
    print(col)

print("\nPreview")
print(df[columns].head())

df.to_csv(
    "../data/cleaned/indian_ecommerce_cleaned.csv",
    index=False
)

print("\nFeature Engineering Completed")
print("Updated dataset saved successfully.")
```
